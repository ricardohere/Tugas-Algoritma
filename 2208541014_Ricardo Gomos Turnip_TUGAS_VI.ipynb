# 5.22.1 Grafik sederhana dengan matplotlib 

```python
#memanggil data dengan format csv
data = pd.read_csv('bunga_iris.csv')
data
```

```python
#mengidentifikasi setiap variabel
sepallength = data['sepal length (cm)']
petallength = data['petal length (cm)']
sepalwidth = data['sepal width (cm)']
petalwidth = data['petal width (cm)']
#berfungsi untuk membuat plot
plt.plot(sepallength,petallength)
#berfungsi untuk memberi judul pada grafik
plt.title ('Grafik bunga iris')
```

# 5.22.2 Properti pembuatan grafik dengan matplotlib 

```python
data
```

```python
x = data['sepal length (cm)']
y = data['petal length (cm)']
x1 = data['sepal width (cm)']
y1 = data['petal width (cm)']

plt.figure(figsize=(12,12), dpi=400) 
#figure digunakan untuk menggambarkan bentuk koordinatnya
#fungsi figsize untuk mengatur ukuran grafik
#dpi berfungsi untuk resolusi gambar

plt.plot(x,y, label = 'Garis Length',color='red',linewidth =3,marker='+',markersize=12,markeredgecolor='black',linestyle='-')
plt.plot(x1,y1, label = 'Garis Width',color='blue',linewidth=3,marker='+',markersize=12,markeredgecolor='yellow',linestyle='-')
#label digunakan untuk menamai linier
#line width adalah untuk ketebalan garisnya
#marker digunakan untuk menentukan titik yang dituju
#markersize digunakan untuk menentukan ukuran pada marker
#markeredgecolor memiliki fungsi untuk memberi warna pada pinggiran marker
#linestyle berfungsi untuk menentuka jenis garis

plt.title('Grafik Bunga Iris', fontdict = {'fontname' : 'Calibri' , 'fontsize' : 12})
#title untuk meberikan judul grafik
plt.xlabel('X', fontdict={'fontname' : 'Arial' , 'fontsize' : 12})
#x label menamakan garis sumbu X
plt.ylabel('Y', fontdict={'fontname' : 'Arial' , 'fontsize' : 12})
#y label menamakan garis sumbu Y
#fontname untuk menentukan jenis font
#fontsize untuk menentukan ukuran dont

plt.xticks(data['sepal length (cm)'])
plt.yticks(data['petal length (cm)'])
plt.legend(loc='upper center')

plt.xticks(data['sepal width (cm)'])
plt.yticks(data['petal width (cm)'])
plt.legend(loc='upper center')

plt.savefig('grafik bunga iris.png', dpi=400)
#savefig digunakan untuk menyimpan grafik pada folder
plt.show()
#show untuk menunjukan grafik
```
